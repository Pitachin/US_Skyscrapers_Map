```python
import geopandas as gpd
import folium
import mapclassify
from matplotlib import pyplot as plt
```

```python
file = gpd.read_file("Distribution_Across_USA.xlsx")
```

```python
file.head() # verify and visualize the file
```

```python
usa = gpd.read_file("https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json")

merged = usa.merge(file, left_on="name", right_on="State", how="left")

merged["Count"] = merged["Count"].fillna(0)

m = merged.explore(
    column = "Count",                
    cmap = "YlOrRd",  # color gradient        
    legend = True,    # show color legend
    tooltip = ["name", "Count"],   # show popup with the 'state' and 'count'   
    tiles = "CartoDB positron",     # base style
    categorical=True, # discrete classes
    style_kwds = dict(opacity=0.9, color="black", weight=0.5),   # arguments for style
    name="Skyscrapers by State"
)
for _, row in merged.iterrows():
    point = row.geometry.representative_point()
    count_text = str(int(row["Count"]))
    folium.map.Marker(
        [point.y, point.x],
        icon=folium.DivIcon(
            html=f"""<div style="font-size:10pt; font-weight:bold; color:black; text-align:center;">{count_text}</div>"""
        ),
    ).add_to(m)
    
m.save("US_Skyscrapers_Explore.html")
m
```
